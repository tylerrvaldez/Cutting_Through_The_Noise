## ***Spotify Dataset Prep for cleaning dataset to use for analysis.***

1) Importing the csv file to be read and converted into a pandas dataframe.

```python
import pandas as pd
```

```python
path_to_csv = '/content/SpotifyFeatures.csv'
csv = pd.read_csv(path_to_csv)
df = pd.DataFrame(csv)
df
```

2) Removing columns that won't be needed for this analysis by dropping the rows and creating a new dataframe.

```python
df1 = df.drop(columns = ['acousticness', 'loudness', 'liveness','speechiness','instrumentalness'])
df1
```

3) Using python, we check for any values in the rows that are duplicates. We have some track name duplicates so we will remove those and just keep the first occurence.

```python
df1 = df1.drop_duplicates(subset = ["track_id"], keep=False)
ids = df1["track_id"]
dup_df = df1[ids.isin(ids[ids.duplicated()])].sort_values("track_id")
dup_df
```

4) Checking for any null values in dataframe before exporting cleaned dataframe for analysis.

```python
print(pd.isnull(df1).sum())
```

5) Converting database to CSV file to be read and queried in SQL.

```python
df1.to_csv('/content/SpotifyFeaturesCleaned.csv')
```
